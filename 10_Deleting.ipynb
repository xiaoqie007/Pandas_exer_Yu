```python
import pandas as pd 
import numpy as np
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = pd.read_csv(url)

iris.head()
```

```python
# create columns for the dataset
iris.columns = ['sepal_length','sepal_width', 'petal_length', 'petal_width', 'class']
iris.head()
```

```python
# is there any missing value in the dataframe?
pd.isnull(iris).sum()
#nice no missing value
```

```python
# Let set the values of the rows 10 to 29 of the column 'petal_length' to NaN
iris.iloc[10:30, 2:3] = np.nan
iris.head(31)
```

```python
# Good, now lets substitute the NaN values to 1.0
iris.petal_length.fillna(1, inplace = True)
iris.head(20)
```

```python
# Now let's delete the column class
del iris['class']
iris.head()
```

```python
#set the first3 rows as NaN
iris.iloc[0:3, :] = np.nan
iris.head()
```

```python
#delete the rows that have NaN
#iris = iris.dropna(how = 'any')
iris = iris.dropna()
iris.head()
```

```python
#pd.isna(iris).sum()
pd.isnull(iris).sum()
```

```python
iris.head()
```

```python
#reset the index so it gegins with 0 again
#iris = iris.reset_index(drop = True)
iris = iris.reset_index()
iris.head()
```

```python
import pandas as pd 
import numpy as np
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
wine = pd.read_csv(url)

wine.head()
```

```python
# Delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns
#wine = wine.drop(wine.columns[[0, 3, 6, 8, 11, 12, 13]])
#!!!KeyError: "['1' '2.43' '2.8' '.28' '1.04' '3.92' '1065'] not found in axis"
wine = wine.drop(wine.columns[[0, 3, 6, 8, 11, 12, 13]], axis= 1)
wine.head()
```

```python
# assign the columns as below:
wine.columns = ['alcohol', 'malic_acid', 'alcalinity_of_ash', 'magnesium', 'flavanoids', 'proanthocyanins', 'hue']
wine.head()
```

```python
# set the values of the first 3 rows from alcohol as NaN
wine.iloc[:3, 0] = np.nan
wine.head()
```

```python
# Now set the value of the rows 3 and 4 of magnesium as NaN
wine.iloc[2:4, 3] = np.nan
wine.head()
```

```python
#fill the value of Nan with the number 10 in alcohol and 100 in magnesium
wine.alcohol.fillna(10, inplace = True)
wine.magnesium.fillna(100, inplace = True)
wine.head()
```

```python
#count the number of missing values
wine.isna().sum()
```

```python
#create an array of 10 random number up until 10
random = np.random.randint(10, size= 10)
random
```

```python
#Use random numbers you generated as an index and assign NaN value to each of cell
wine.alcohol[random] = np.nan
wine.head(10)
```

```python
random1 = np.random.randint(10, size= 15)
random1
```

```python
wine.alcalinity_of_ash[random1] = np.nan
wine.head()
```

```python
wine.isnull().sum()
```

```python
#delete the rows that contain missing values #dropna
wine = wine.dropna(axis= 0, how= 'any')
wine = wine.dropna(axis = 0)
wine.head()
```

```python
# print only the non-null values in alcohol
mask = wine.alcohol.notnull()
mask.head(10)
```
