### Wind Statistics

```python
import pandas as pd
import datetime
```

```python
data_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data'
```

```python
# parse_dates gets 0,1,2 columns and parse them as the index
data = pd.read_table(data_url, sep = "\s+", parse_dates = [[0,1,2]]) 
data.head()
```

```python
# Year 2061? Do we really have data from this year? Create a function to fix it and apply it.
#the problem is that the dates are 2061 and so on...
#function that uses datetime

def fix_century(x):
    year = x.year - 100 if x.year > 1989 else x.year
    return datetime.date(year, x.month, x.day)

#apply the function fix_century on the column and replace the values to the right ones
data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix_century)
```

```python
data.info()
```

```python
data.head()
```

```python
# set the right dates as the index.pay attention at the data type, it should be datetime64[ns].

#transfrom Yr_Mo_Dy it to datetype datetime64
data["Yr_Mo_Dy"] = pd.to_datetime(data["Yr_Mo_Dy"])

#set 'Yr_Mo_Dy' as the index
data = data.set_index('Yr_Mo_Dy')

data.head()
```

```python
#compute how many values are missing for each location over the entire record
data.isnull().sum()
#they should be ignored in all calculations below.
```

```python
#computer how many non-missing values there are in total.

#data.notnull().sum()

#OR
#number of columns minus the number of missing values for each location
data.shape[0] - data.isnull().sum()
```

```python
#calculation the mean windspeeds of the windspeeds over all the locations and all the times.
data.fillna(0).values.flatten().mean()
```

```python
# create a DataFrame called loc_stats and calculate the min , max and mean windspeeds 
#and standard deviations of the windspeeds at each location over all the days
#data.describe()
data.describe(percentiles=[])
```

```python
#create the dataframe
day_stats = pd.DataFrame()
day_stats
```

```python
#this time we determine axis equals to one so it get each row.
day_stats['min'] = data.min(axis = 1)
day_stats['max'] = data.max(axis = 1)
day_stats['mean'] = data.mean(axis = 1)
day_stats['std'] = data.std(axis = 1)
day_stats.head()
```

```python
#不设置axis = 1时，的区别
day_stats1 = pd.DataFrame()
day_stats1['min'] = data.min()
day_stats1['max'] = data.max()
day_stats1['mean'] = data.mean()
day_stats1['std'] = data.std()
day_stats1.head()
```

```python
#find the average windspeed in January for each location.
data.loc[data.index.month == 1].mean()
```

```python
# downsample the record to a yearly frequency for each location.
data.groupby(data.index.to_period('Y')).mean()
```

Series.dt.to_period(self, *args, **kwargs)[source]

Cast to PeriodArray/Index at a particular frequency.
Converts DatetimeArray/Index to PeriodArray/Index.
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.dt.to_period.html

```python
# downsample the record to a monthly frequency for each loaction
data.groupby(data.index.to_period('M')).mean().tail()
```

```python
#downsample the record to a weekly frequency for each location.
data.groupby(data.index.to_period('W')).mean().head()
```

```python
#Calculate the min, max and mean windspeeds and standard deviations of the windspeeds 
#across all locations for each week (assume that the first week starts on January 2 1961) 
#for the first 52 weeks
#data.groupby(data.index.to_period('W')).describe().head()#这个方法不太准确

#关注resample方法：和agg方法：
#resample data to 'W' week and use the functions
weekly = data.resample('W').agg(['min', 'max', 'mean', 'std'])

#slice it for the first 52 weeks and locations
weekly.loc[weekly.index[1:53], 'RPT':'ROS'].head()
```
